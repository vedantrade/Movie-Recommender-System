```python
import pandas as pd
import numpy as np
df_credits = pd.read_csv('/content/drive/MyDrive/Movie recommender system/tmdb_5000_credits.csv')
df_movies = pd.read_csv('/content/drive/MyDrive/Movie recommender system/tmdb_5000_movies.csv')
```

```python
df_credits = df_credits.drop(['title'],axis = 1)
```

```python
df_credits.columns = ['id','cast','crew']
```

```python
df_credits.head()
```

```python
df_movies = df_movies.merge(df_credits,on = 'id')
df_movies.head()
```

```python
c = df_movies['vote_average'].mean()
m = df_movies['vote_count'].quantile(q=0.9)
print(c)
print(m)
```

```python
demo_movies = df_movies.copy().loc[(df_movies['vote_count']>m)]
demo_movies.shape
```

```python
def wr(x):
  v = x['vote_count']
  r = x['vote_average']
  w = ((v/(v+m))*r)+((m/(v+m))*c)
  return w
```

```python
wr(df_movies)
```

```python
demo_movies['weighted_rating'] = wr(df_movies)
```

```python
demo_movies.head()
```

```python
demo_movies = demo_movies.sort_values(by = 'weighted_rating', ascending = False)
```

## **Content based filtering**

```python
df_movies['overview'].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer as tfidvector
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
tfidf = tfidvector(stop_words = 'english')
```

```python
df_movies['overview'].isnull().values.any()
```

```python
df_movies['overview'] = df_movies['overview'].fillna('')
```

```python
df_movies['overview'].isnull().values.any()
```

```python
matrix = tfidf.fit_transform(df_movies['overview'])
```

```python
matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity,linear_kernel
```

```python
cos_matrix = cosine_similarity(matrix)
```

```python
cos_matrix
```

```python
indices = pd.Series(df_movies.index,index = df_movies['title'])
indices
```

```python
def get_recommendation(title):
  index = indices[title]
  similarity = list(enumerate(cos_matrix[index]))
  similarity = sorted(similarity,key = lambda x:x[1],reverse = True)
  movie_list = []
  for i in similarity:
    movie_list.append(i[0])
  movie_list = movie_list[1:11]
  return df_movies['title'].loc[movie_list]
```

```python
get_recommendation('The Dark Knight')
```

```python
df_movies['keywords'].head()
```

```python
from ast import literal_eval
```

```python
col = ['cast','crew','keywords','genres']
```

```python
for i in col:
  df_movies[i] = df_movies[i].apply(literal_eval)
```

```python
df_movies['cast'].head()
```

```python
def find_director(crew):
  for i in crew:
    if i['job'] == 'Director':
      return i['name']
```

```python
df_movies['directors'] = df_movies['crew'].apply(find_director)
```

```python
df_movies[['title', 'cast', 'keywords', 'genres','directors']]
```

```python
def top_3(x):
  if isinstance(x,list):
    names = [i['name'] for i in x]
    if len(names)>3:
      names = names[:3]
    return names
```

```python
col_2 = ['cast','keywords','genres']
for j in col_2:
  df_movies[j] = df_movies[j].apply(top_3)
```

```python
df_movies[['title', 'cast', 'keywords', 'genres','directors']]
```

```python
def clean_data(y):
  temp = []
  for i in y:
    temp.append(str.lower(i.replace(' ','')))
  return temp
```

```python
for j in col_2:
  df_movies[j] = df_movies[j].apply(clean_data)
```

```python
df_movies.head()
```

```python
def create_soup(x):
  return ' '.join(x['keywords'])+' '.join(x['genres'])+' '.join(x['cast'])
```

```python
df_movies['soup'] = df_movies.apply(create_soup, axis = 1)
```

```python
count = CountVectorizer(stop_words = 'english')
```

```python
count_matrix = count.fit_transform(df_movies['soup'])
```

```python
cos_matrix1 = cosine_similarity(count_matrix)
```

```python
indices1 = pd.Series(df_movies.index,index = df_movies['title'])
```

```python
def get_recommendation1(title, cosine_similarity = cos_matrix1):
  index = indices1[title]
  similarity = list(enumerate(cos_matrix1[index]))
  similarity = sorted(similarity,key = lambda x:x[1],reverse = True)
  movie_list = []
  for i in similarity:
    movie_list.append(i[0])
  #movie_list = movie_list[1:11]
  return df_movies['title'].loc[movie_list]
```
